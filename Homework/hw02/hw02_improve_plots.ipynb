```python
# Import packages we'll need for the plots. You don't have to modify this cell.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

### Dataset 1: Strength vs. Time

```python
# download dataset 1
!wget 'https://raw.githubusercontent.com/ryantrainor/PHY322/refs/heads/main/Homework/hw02/hw02-data1.csv' -O hw02-data1.csv
```

```python
# read data from csv file
df1 = pd.read_csv('hw02-data1.csv')

# create figure object and plot both sets of data with error bars
fig,ax=plt.subplots()
ax.errorbar(df1.xa,df1.ya,df1.ea,ecolor='k',ls='none',elinewidth=0.5,zorder=1,
            marker='o',mfc='red',ms=5,mec='none',label='Group A')
ax.errorbar(df1.xb,df1.yb,df1.eb,ecolor='k',ls='none',elinewidth=0.5,zorder=1,
            marker='o',mfc='tab:red',ms=5,mec='none',label='Group B')

# set properties of axes, ticks, and labels
ax.set_xlabel('Time (s)',size=6)
ax.set_ylabel('Strength (N)',size=6)
ax.axis([-15,15,0,3])
ax.tick_params(axis='both',labelsize=6)
ax.xaxis.set_major_locator(MultipleLocator(10))
ax.yaxis.set_major_locator(MultipleLocator(1))

# create legend
ax.legend(fontsize=6)

# automatically adjust figure layout and save figure
fig.tight_layout()
```

### Dataset 2: Mass histogram

```python
# download dataset 2
!wget 'https://raw.githubusercontent.com/ryantrainor/PHY322/refs/heads/main/Homework/hw02/hw02-data2.csv' -O hw02-data2.csv
```

```python
# read data from csv file
df2 = pd.read_csv('hw02-data2.csv')

# create figure object, define bins, and plot histogram
fig,ax=plt.subplots()
bins=np.arange(0,50,0.5)
ax.hist(df2.x,bins=bins,facecolor='purple',edgecolor='dodgerblue',
        hatch='//',fill=True,lw=1.5)

# set properties of axes, ticks, and labels
ax.set_xlabel('Mass (kg)',size=12)
ax.set_ylabel('Number',size=12)
ax.set_xlim([0,50])
ax.tick_params(axis='both',labelsize=10)
ax.tick_params('both',which='both',top=True,right=True,labelright=True,labeltop=True)
ax.xaxis.set_major_locator(MultipleLocator(2))
ax.xaxis.set_minor_locator(MultipleLocator(0.5))
ax.yaxis.set_major_locator(MultipleLocator(0.25))
ax.yaxis.set_minor_locator(MultipleLocator(0.05))

# automatically adjust figure layout and save figure
fig.tight_layout()
```

### Dataset 3: Force vs. Position

```python
# download datasets 3a, 3b, 3c
!wget 'https://raw.githubusercontent.com/ryantrainor/PHY322/refs/heads/main/Homework/hw02/hw02-data3a.csv' -O hw02-data3a.csv
!wget 'https://raw.githubusercontent.com/ryantrainor/PHY322/refs/heads/main/Homework/hw02/hw02-data3c.csv' -O hw02-data3b.csv
!wget 'https://raw.githubusercontent.com/ryantrainor/PHY322/refs/heads/main/Homework/hw02/hw02-data3c.csv' -O hw02-data3c.csv
```

```python
# read data from csv files
df3a = pd.read_csv('hw02-data3a.csv')
df3b = pd.read_csv('hw02-data3b.csv')
df3c = pd.read_csv('hw02-data3c.csv')

# create figure object and plot data with error bars
fig,ax=plt.subplots()
ax.errorbar(df3a.x,df3a.y,df3a.e,ecolor='k',ls='none',elinewidth=0.5,zorder=1,
            marker='o',mfc='k',ms=3,mec='none',label='Data 1')
ax.errorbar(df3b.x,df3b.y,df3b.e,ecolor='k',ls='none',elinewidth=0.5,zorder=1,
            marker='*',mfc='k',ms=8,mec='none',label='Data 2')
ax.errorbar(df3c.x,df3c.y,df3c.e,ecolor='k',ls='none',elinewidth=0.5,zorder=1,
            marker='>',mfc='k',ms=5,mec='none',label='Data 3')

# define linear function to use for model line
def linear_func(x):
    return 22*x+232

# draw model line
xfit=np.linspace(0,100,100)
ax.plot(xfit,linear_func(xfit),'k--',lw=0.5,label='Fit')

# set properties of axes, grid, ticks, and labels
ax.set_xlabel('Foorce (N)',size=12)
ax.set_ylabel('Potision (cm)',size=12)
ax.axis([0,100,0,2800])
ax.grid(which='major',axis='both',color='k',lw=2)
ax.grid(which='minor',axis='both',color='k',lw=1.5)
ax.tick_params(axis='both',which='major',labelsize=10,width=2)
ax.tick_params(axis='both',which='minor',width=1.5)
ax.tick_params(axis='both',which='both',top=True,right=True)
for spine in ax.spines.values():
    spine.set_linewidth(2)
ax.xaxis.set_major_locator(MultipleLocator(20))
ax.xaxis.set_minor_locator(MultipleLocator(2))
ax.yaxis.set_major_locator(MultipleLocator(500))
ax.yaxis.set_minor_locator(MultipleLocator(50))

# automatically adjust figure layout and save figure
fig.tight_layout()
```
